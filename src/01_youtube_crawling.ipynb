```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time, pandas as pd
import chromedriver_autoinstaller

# 크롬 옵션 설정
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')  # GUI 비활성화
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('lang=ko_KR')

# 드라이버 자동설치 및 실행
chromedriver_autoinstaller.install()
driver = webdriver.Chrome(options=chrome_options)

# (필요시 URL 지정)
url = "https://www.youtube.com/watch?v=Fd_bsWBqlho"
driver.get(url)
```

```python

#코랩에서 driver 설정 그대로 가져오기

# Open the YouTube video page
driver.get(url)

# Wait for the page to load
time.sleep(5)

# Scroll down to load more comments
body = driver.find_element(By.TAG_NAME, 'body')
num_of_pagedowns = 20  # You can adjust this value based on the number of comments you expect
while True:
    last_height = driver.execute_script("return document.documentElement.scrollHeight")
    for _ in range(num_of_pagedowns):
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(1.5)  # Adjust the sleep time if needed

    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Find all comment elements
comments = driver.find_elements(By.CSS_SELECTOR, 'yt-attributed-string[slot="content"]')

# Extract the text of each comment
comments_text = [comment.text for comment in comments]


# Save to a CSV file
df = pd.DataFrame(comments_text, columns=['Comment'])
df.to_csv('youtube_comments.csv', index=False, encoding='utf-8-sig')
driver.quit()


print("Comments saved to youtube_comments27.csv")

```
