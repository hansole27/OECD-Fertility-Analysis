```python
#특수문자, 숫자 제거하기
import re

def del_noise(df):

  cleaned_art = pd.DataFrame()

  # 문자와 공백을 제외한 모든 특수 문자 및 숫자 제거
  for i in range(len(df)):

    cleaned_art.loc[i, 'Comment'] = re.sub(r'[^A-Za-z가-힣\s]', '', df.loc[i, 'Comment'])
  return cleaned_art

df1 = del_noise(df)
```

```python
#명사 추출하기
from konlpy.tag import Okt

# Okt 형태소 분석기 객체 생성
okt = Okt()

def extract_nouns(text):
    if isinstance(text, str):  # 입력이 문자열인지 확인
        nouns = okt.nouns(text)
        filtered_nouns = [noun for noun in nouns if len(noun) >= 2]  # 길이가 2 이상인 명사만 추출
        return filtered_nouns
    else:
        return []  # 문자열이 아닌 경우 빈 리스트 반환

# '내용' 열에서 명사를 추출하여 'noun' 열에 추가
df1['noun'] = df1['Comment'].apply(extract_nouns)
df1

```

```python
#명사 추출이 리스트 형태로 되며 생긴 '[,]와 같은 기호 추가 제거
for q in range(len(df1)):
  d = df1['noun'][q]
  # Join the list elements into a string using an empty space as the separator
  d = ' '.join(d)
  d = d.replace(",","")
  d= d.replace("'",'')
  d= d.replace("[",'')
  d= d.replace("]",'')

  #명사만 존재하는 nouns 컬럼에 저장
  df1.loc[q,'nouns'] = d
df1
```

```python
#불용어 제거1
#불용어 정의(대~충 훑어보고 적절히 추가해주기..)
stoplist=[
    '아직도', '내가', '아이', '없이', '인구', '대한민국', '나라가', '절대', '보면', '없고', '이렇게',
    '계속', '민주당', '낳으면', '할', '애', '뭐', '합니다', '하는', '허경영', '얼마나', '때문에', '애들',
     '된다', '없는', '이런', '근데', '못', '다른', '말고', '것',  '있는데', '있다', '하면',
    '년', '정책을',  '보고', '일',   '우리', '진짜', '어차피',  '또', '그냥',
    '이제', '낳고', '모든', '아닌가', '누가', '그리고', '그게', '그럼', '대한', '있는', '너무', '안',
   '무슨', '모두', '제대로', '사람',  '참', '없다', '나라', '이', '국민', '무조건',
    '많이', '좋은', '하지', '같은', '위해', '좀', '아니고', '아니라', '억', '하고',  '뭔', '이재명',
    '다', '해도', '정말', '잘', '한', '명', '위한', '많은', '해라', '더', '수', '사람이', '그렇게', '아이를',
    '어떻게',  '이미',  '하나',  '그', '애를', '가장', '사람들이',  '왜', '제발',
    '나라를', '이게', '문제가', '한다', '만원', '지금', '윤석열', '결국','직구','기자','기사','매우','별로','항상','정말','예전','까까좃다구','좀더','생각','정도','소리',
    '때문','한동훈','이유','누구','좌파','문재인','가지','시작','이준석','머리','얘기'

]

stopw = pd.DataFrame(stoplist)
stopw

for q in range(len(stopw)):
  stopw.loc[q,'word'] = stopw.loc[q,0].replace(',','')
stopw= pd.DataFrame(stopw['word'])

```

```python
#불용어 제거2
# 불용어의 word 컬럼 헤더에 공백이나 특수 문자가 포함된 것 같으니, 첫 번째 컬럼을 명시적으로 사용
stop_words_set = set(stopw.iloc[:, 0].str.strip())  # Stripping any whitespace

# 문자열에서 불용어를 필터링하는 함수 정의
def filter_stop_words(nouns_str):
    # 문자열을 단어로 분리
    words = nouns_str.split()
    # 불용어 제거
    filtered_words = [word for word in words if word not in stop_words_set]
    # 단어를 다시 문자열로 결합
    return ' '.join(filtered_words)
```

```python
# 'nouns' 컬럼에 함수 적용하고 필터링된 명사로 새 컬럼 생성
df1['nouns'] = df1['nouns'].apply(filter_stop_words)
df1

#tf-idf 활용해서 최종적으로 단어별 빈도수 확인
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df1['nouns'])

# 단어별 TF-IDF 값을 데이터프레임으로 변환
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# 각 단어의 빈도수 계산
word_count_vectorizer = TfidfVectorizer(use_idf=False, norm=None)
word_count_matrix = word_count_vectorizer.fit_transform(df1['nouns'])
word_count_df = pd.DataFrame(word_count_matrix.toarray(), columns=word_count_vectorizer.get_feature_names_out())

# 모든 문서에서의 단어 빈도수 합산
word_frequencies = word_count_df.sum(axis=0).sort_values(ascending=False).reset_index()
word_frequencies.columns = ['word', 'frequency']

# 결과 출력
print("단어 빈도수 데이터프레임:")
print(word_frequencies)
```

```python
word_frequencies.to_csv('word_frequencies.csv', index=False)
```
