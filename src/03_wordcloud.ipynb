```python
# 런타임 다시 시작 후  이부분 실행
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
#
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_name = mpl.font_manager.FontProperties(fname=path).get_name()
plt.rcParams['font.family'] = font_name
```

```python
# 워드클라우드 코드
# 위에 word_frequencies 데이터프레임 그대로 사용해서 시각화 가능합니다

import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive') # Mount your Google Drive

# CSV 파일 로드
word_frequencies = '/content/drive/MyDrive/word_frequencies1.csv'
word_frequencies_df = pd.read_csv(word_frequencies)

# 워드클라우드를 생성하기 위한 단어와 빈도수 딕셔너리 생성
word_freq_dict = dict(zip(word_frequencies_df['word'], word_frequencies_df['frequency']))

# 워드클라우드 생성
wordcloud = WordCloud(
    width=1400,
    height=700,
    background_color='white',
    font_path='/usr/share/fonts/truetype/nanum/NanumGothic.ttf',  # 여기서 글꼴 경로를 설정할 수 있습니다
    collocations=False  # 어구가 아닌 단어 단위로 분석하도록 설정
).generate_from_frequencies(word_freq_dict)

# 워드클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 축 제거
plt.show()
```
